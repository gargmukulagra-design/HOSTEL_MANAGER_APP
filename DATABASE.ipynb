```python
import pandas as pd 
import sqlite3
```

```python
conn = sqlite3.connect("Students.DB")
client = conn.cursor()
```

```python
#SQL COMMAND
query = """CREATE TABLE IF NOT EXISTS timetable(
    TIME TEXT,
    MONDAY TEXT,
    TUESDAY TEXT,
    WEDNESDAY TEXT,
    THURSDAY TEXT,
    FRIDAY TEXT,
    SATURDAY TEXT,
    SUNDAY TEXT)
    """
```

```python
client.execute(query)
```

```python
conn.commit()
```

```python
#timetable
rows = [

    ('9-10', 'DBMS', 'ML', 'OOPS', 'DBMS LAB', 'FREE', 'SPORTS', 'HOLIDAY'),

    ('10-11', 'OOPS', 'DBMS', 'ML', 'FREE', 'ML LAB', 'FREE', 'HOLIDAY'),

    ('11-12', 'ML', 'OOPS', 'DBMS', 'ML LAB', 'DBMS', 'FREE', 'HOLIDAY')

]
```

```python
client.execute("DELETE FROM timetable")
conn.commit()


#inserting the data 
for row in rows:

    client.execute("""
    INSERT INTO timetable
    VALUES (?, ?, ?, ?, ?, ?, ?, ?)
    """, row)
```

```python
timetable = pd.read_sql_query("""
                                SELECT * from timetable
                                """,conn)
```
